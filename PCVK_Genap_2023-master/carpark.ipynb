Hasil dari penerapan metode Sobel edge detection pada suatu gambar yang terdiri dari 3 contoh sample gambar

```python
# Sobel Edge Detection

img1 = cv.imread('/content/drive/MyDrive/PCVK/car-park/picture5.jpg') 

# converting to gray scale
gray = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

# remove noise
img2 = cv.GaussianBlur(gray,(7,7),0)


# obtaining a horizontal and vertical Sobel filtering of the image
img_sobelx = cv.Sobel(img1, cv.CV_64F, 1, 0, ksize=3)
img_sobely = cv.Sobel(img1, cv.CV_64F, 0, 1, ksize=3)

# image with both horizontal and vertical Sobel kernels applied
img_sobelxy = cv.addWeighted(cv.convertScaleAbs(img_sobelx), 0.5, cv.convertScaleAbs(img_sobely), 0.5, 0)

f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
axarr[0].imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(img_sobelxy,cmap = 'gray')
```

Lalu yang terakhir ada penggunaan metode cascade classifier pada 3 contoh gambar berbeda untuk memuncul kan rectangle

```python
cascade = cv.CascadeClassifier('/content/drive/MyDrive/PCVK/haarcascades/cars.xml')
car = cv.imread('/content/drive/MyDrive/PCVK/car-park/picture5.jpg')
car1 = cv.imread('/content/drive/MyDrive/PCVK/car-park/picture5.jpg')

jumlah =0;
gray = cv.cvtColor(car, cv.COLOR_BGR2GRAY)
cars = cascade.detectMultiScale(gray)
for(x,y,w,h) in cars: 
 cv.rectangle(car,(x,y),(x+w,y+h),(0,255,0),5) 
 jumlah+=1;
plt.figure(figsize = (10,10)) 

f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(car, cv.COLOR_BGR2RGB))
axarr[0].imshow(cv.cvtColor(car, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(car1, cv.COLOR_BGR2RGB))
axarr[1].imshow(cv.cvtColor(car1, cv.COLOR_BGR2RGB))
plt.show()
print(jumlah)
```
